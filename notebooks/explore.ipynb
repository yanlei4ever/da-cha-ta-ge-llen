```python
from datetime import datetime, timedelta
from collections import defaultdict
from math import ceil
import pandas as pd

rel_data = pd.read_csv('../data/raw/train_2011_2012_2013.csv', sep=';',usecols=['DATE','ASS_ASSIGNMENT','CSPL_RECEIVED_CALLS'])
aggregated = rel_data.groupby(['DATE', 'ASS_ASSIGNMENT']).sum().reset_index()

pairs = defaultdict(int)

for i,row in aggregated.iterrows():
    date, ass, count = row['DATE'], row['ASS_ASSIGNMENT'], row['CSPL_RECEIVED_CALLS']
    pairs[(date, ass)] = count

def final_model(dateString, ass):
    date = datetime.strptime(dateString[:-4], '%Y-%m-%d %H:%M:%S')
    
    before = str(date - timedelta(days=7)) + '.000'
    after = str(date + timedelta(days=7)) + '.000'
    
    val_before = pairs[(before, ass)]
    val_after = pairs[(after, ass)]
    
    #print val_before, val_after
    return ceil((val_before+val_after)/2.)
        
        
def submit(filepath):
    """ Read submission file and make prediction, saving output to new file. """

    with open(filepath,'r') as submit, open('output.txt','w') as output:

        # copy heading to file
        output.write(submit.readline())

        for line in submit:
            datetime, ass_assignment, _ = line.split('\t')

            prediction = final_model(datetime, ass_assignment)

            output.write(output_line.format(datetime=datetime,
                                            ass_assignment=ass_assignment,
                                            prediction = prediction))

output_line = "{datetime}\t{ass_assignment}\t{prediction}\r\n"       
submit('submission.txt')
```
